```python
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
x=np.linspace(0,4,100)
y=np.log(x)

f, axarr = plt.subplots(2, sharex=True)
f.suptitle('Forex v. Money Market')

axarr[0].plot(x, -y+2,color='forestgreen',label='broo')
axarr[1].plot(x, y-2,color='firebrick',label='bro')
axarr[0].axvline(1.5,label='aoo')
axarr[1].axhline(-1.6,label='jdn')
axarr[1].legend(loc=0)
axarr[0].legend(loc=0)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=2, cols=1,
                    shared_xaxes=True,
                    vertical_spacing=0.01)

fig.add_trace(go.Line(x=x,y=-y+2,mode='lines',name='$E_{CHF/€}\\bigl(R_{€},E^e_{CHF/€}\\bigr)$',
                      line_color='deepskyblue',hovertemplate='Expected € Return (in CHF)<extra></extra>'),row=1,col=1)
fig.add_vline(1.5,row=1,col=1,
              annotation_text='Swiss Interest Rate')


fig.add_trace(go.Line(x=x,y=y-2,mode='lines',name='$-L\\bigl( R_{CHF}, Y_{\\text{Swiss}} \\bigr)$',
                      line_color='steelblue',
                      hovertemplate='Demand for Money<extra></extra>'),row=2,col=1)
fig.add_hline(-1.6,row=2,col=1,name='$\\frac{M_{CHF}}{P}$',line_color = 'darkblue',
              annotation_text='Money Supply',annotation_position='bottom right')
fig.add_vline(1.5,row=2,col=1,line_dash='dot')

fig.update_layout(height=350, width=400,margin=dict(l=60, r=0, t=30, b=50),template='plotly_white')
fig.update_xaxes(title_text="$R_{CHF}$", row=2, col=1)
fig.update_yaxes(title_text="$E_{CHF/€}$", row=1, col=1)
fig.update_yaxes(title_text="$-M_{CHF}/P$", row=2, col=1)

fig.update_layout(legend=dict(
    orientation="h",
    yanchor="bottom",
    y=1.02,
    xanchor="right",
    x=1
))

fig.add_trace(go.Scatter(x=[1.5],y=[1.6],mode='markers+text',marker_color='black',
                         showlegend=False,marker=dict(size=8),text='$E_{\\text{Forex}}$',textposition='top right'),row=1,col=1)
fig.add_trace(go.Scatter(x=[1.5],y=[-1.6],mode='markers+text',marker_color='black',
                         showlegend=False,marker=dict(size=8),text='$E_{MM}$',textposition='bottom right'),row=2,col=1)

fig.add_hline(0,line_width=0.8,row=1,col=1)

#,title_text="Forex Market & Money Market Equilibria"
fig.show()
```
